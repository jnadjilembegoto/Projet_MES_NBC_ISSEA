```python
import pandas as pd 
def tab_large_en_long(data,debut_year,fin_year):
    base_panel=pd.melt(data, id_vars=['Country','Variable'], value_vars=[i for i in range(debut_year,fin_year+1,1)], var_name='Year',value_name='valeur')
    base_panel['Year']=base_panel['Year'].astype(int)

    # Convertir `valeur` en num√©rique et remplacer les erreurs par NaN
    base_panel['valeur'] = pd.to_numeric(base_panel['valeur'], errors='coerce')

    base_panel=base_panel.pivot_table(index=['Country','Year'],columns='Variable', values='valeur').reset_index()
    return base_panel
```

```python
def load_data(sheet_name):
    return pd.read_excel("Datas/ml_wdi_data6.xlsx",sheet_name=sheet_name)

base = load_data("INSCRIPTION")
```

```python
base.head()
```

```python
#base.isnull().sum()
```

```python
data=base.drop(columns=['Variables', 'Pays'],axis=1,inplace=False)
data.head(5)
```

```python
# Fonction pour remplacer les NaN par la moyenne de la ligne
def remplir_par_moyenne(row):
    moyenne = row.mean()  # Calcul de la moyenne des valeurs non manquantes de la ligne
    return row.fillna(moyenne)  # Remplacer les NaN par cette moyenne

# Appliquer la fonction √† chaque ligne
data = data.apply(remplir_par_moyenne, axis=1)
data.head()
```

```python
#nouvelle base sans les valeurs manquantes
base_finale=pd.concat([base[['Pays','Variables']],data],axis=1)
base_finale.head()
```

```python
def tab_large_en_long(data, debut_year, fin_year):
    # Conversion au format long
    base_panel = pd.melt(
        data, 
        id_vars=['Pays', 'Variables'], 
        value_vars=[str(year) for year in range(debut_year, fin_year + 1)],  # Ann√©es sous forme de colonnes
        var_name='Year', 
        value_name='valeur'
    )
    
    # Conversion des ann√©es en entier
    base_panel['Year'] = base_panel['Year'].astype(int)
```
